##### Analysis and visualzation for non-binary columns

```python
# Creating plots for each non-binary column to visualize its relationship with "Diabetes_binary"

# Set up the figure size and layout for the plots
plt.figure(figsize=(16, 20))

# Loop through each non-binary column and create a plot
for i, column in enumerate(non_binary_columns):
    plt.subplot(4, 2, i + 1)
    # Plot a boxplot to show distribution of the non-binary column for both diabetes and non-diabetes groups
    sns.boxplot(x='Diabetes_binary', y=column, data=data)
    plt.title(f"{column} vs Diabetes_binary")
    plt.xlabel('Diabetes_binary')
    plt.ylabel(column)

# Adjust layout for better visibility
plt.tight_layout()
plt.show()
```

Conclusion from boxplot visualization
1. BMI (Body Mass Index)
- Plot: The boxplot shows that individuals with diabetes generally have a higher BMI compared to those without diabetes. The median BMI for the diabetic group is higher, indicating a positive relationship between higher BMI and diabetes.
- Analysis: This suggests that obesity or being overweight is more common among individuals with diabetes, which aligns with known risk factors for type 2 diabetes.
2. GenHlth (General Health)
- Plot: The general health scores are generally worse for individuals with diabetes. Higher values represent poorer health, and the median value for people with diabetes is higher compared to the non-diabetic group.
- Analysis: Individuals with diabetes tend to report worse general health, which makes sense given the impact of diabetes on overall health.
3. MentHlth (Mental Health)
- Plot: The distribution of mental health days (days with poor mental health) is slightly higher for individuals with diabetes, though there is more variation in both groups.
- Analysis: People with diabetes seem to experience slightly more mental health issues, although the difference is not as pronounced as in other variables.
4. PhysHlth (Physical Health)
- Plot: People with diabetes tend to report more poor physical health days compared to those without diabetes. The median number of poor physical health days is clearly higher for the diabetic group.
- Analysis: Poor physical health is more prevalent among individuals with diabetes, which could be due to the complications of the disease affecting daily life.
5. Age
- Plot: The diabetic group tends to be older, with the median age noticeably higher than that of the non-diabetic group.
- Analysis: Age is a known risk factor for diabetes, especially type 2 diabetes, as the likelihood of developing the condition increases with age.
6. Education
- Plot: There is not much visible difference between the diabetic and non-diabetic groups in terms of education levels. However, the non-diabetic group appears to have a slightly more varied range.
- Analysis: Education doesn't show a strong correlation with diabetes in this dataset, though socio-economic factors often play a role in health outcomes.
7. Income
- Plot: Individuals without diabetes appear to have slightly higher income levels compared to those with diabetes. The median income is lower in the diabetic group.
- Analysis: Lower income could be associated with reduced access to healthcare and a greater prevalence of lifestyle factors that increase the risk of diabetes.

This analysis highlights the relationships between non-binary variables and diabetes, with significant patterns in BMI, general health, physical health, and age.

#### Correlation Analysis

```python
corr = data.corrwith(data['Diabetes_binary'])
# print(corr)
# Filter rows where the correlation is greater than certain percentage
filtered_corr = corr[corr > 0.2]

# Print the result
print(filtered_corr)

# Count the number of rows with corr > 0.2
count_filtered = filtered_corr.count()

# Total number of rows in the correlation series
total_rows = corr.count()

# Calculate proportion
proportion = count_filtered / total_rows

# Print the results
print(f'Rows with corr > 0.2: {count_filtered} out of {total_rows}')
print(f'Proportion: {proportion:.2%}')
```
